```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from utility.utils import json_2_dataframe
from utility.utils import train_test_spliter
from utility.utils import corpus_entity_info
```

```python
data = json_2_dataframe('../data/ChatbotCorpus.json')
splited_data = train_test_spliter(data)
train_data, test_data = splited_data.train, splited_data.test
```

```python
print(f"Train data size: {len(train_data)}")
print(f"Test data size: {len(test_data)}")
```

```python
train_ner = pd.concat([pd.DataFrame(ent) for ent in train_data.entities.values])
test_ner = pd.concat([pd.DataFrame(ent) for ent in test_data.entities.values])
```

###  Entity types within the chatbot corpus

```python
entity_types = corpus_entity_info(train_ner, test_ner)
print(entity_types)
```

### Entity types within the training data

```python
sns.set(rc={'figure.figsize':(11.7,6.27)})
sns.countplot(train_ner['entity'],label="Count",edgecolor=sns.color_palette("dark", 3))
plt.show()
```

### Entity types within the test data

```python
sns.countplot(test_ner['entity'],label="Count",edgecolor=sns.color_palette("dark", 3))
plt.show()
```
